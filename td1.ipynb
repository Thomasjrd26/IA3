## **4. Régression Linéaire (un perceptron)**

- Le but de cet exercice est de faire apprendre à un neurone une fonction affine (y = ax + b). Dans cet exercice nous avons choisi a = 0.6 et b = 2.
- Le neurone (un perceptron) possède une seule entrée (les x) et une seule sortie, la fonction y à laquelle nous avons ajouté du bruit (selon une loi normale).
- Les 2 paramètres que le neurone doit apprendre sont évidemment la pente de la droite (a) et l'ordonnée à l'origine (b).


### Import libraries

````python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
import tensorflow as tf
print (tf.__version__)
````

### Create noisy data

````python
# Parameters (y = a*x + b)
a = 0.6
b = 2

# Create noisy data
x_data = np.linspace(-10, 10, num=10000)
noise = np.random.normal(size=10000)
y_data = a * x_data + b + noise

y_real = a * x_data + b

print('Data created successfully')
````

## **5. Un classifier linéaire**

- Le but de cet exercice est de séparer par une droite 2 classes de points (classe 0 et classe 1).
- Les données étant linéairement séparables (on le voit en les visualisant), un seul neurone peut être utilisé (un perceptron).
- Le problème se ramène donc à une régression linéaire où l’on cherche l’équation de la droite (y=ax+b) séparant les 2 classes de points…

````python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
import tensorflow as tf
````

### Generating two classes of random points in a 2D plane

````python
num_samples_per_class = 1000
negative_samples = np.random.multivariate_normal(
    mean=[0, 3],
    cov=[[1, 0.5],[0.5, 1]],
    size=num_samples_per_class)
positive_samples = np.random.multivariate_normal(
    mean=[3, 0],
    cov=[[1, 0.5],[0.5, 1]],
    size=num_samples_per_class)
````

### Stacking the two classes into an array with shape (2000, 2)

````python
inputs = np.vstack((negative_samples, positive_samples)).astype(np.float32)
````

### Generating the corresponding targets (0 and 1)

````python
targets = np.vstack((np.zeros((num_samples_per_class, 1), dtype="float32"),
                     np.ones((num_samples_per_class, 1), dtype="float32")))
````

````python
targets.shape
````

### Plotting the two point classes

````python
plt.scatter(inputs[:, 0], inputs[:, 1], c=targets[:, 0])
plt.show()
````

### Creating the linear classifier variables

````python
input_dim = 2
output_dim = 1
````

````python
# Create the model
model = keras.Sequential()
# A completer

# Create the model (re-defining it to ensure layers are added)
model = keras.Sequential()
model.add(keras.layers.Dense(units=output_dim, activation='linear', input_shape=(input_dim,)))

# Compilation du modèle (re-compiling it)
model.compile(loss='mse', metrics=['accuracy'], optimizer="adam")
````

````python
# Display the model (only 2 parameters to optimize)
model.summary()
````
